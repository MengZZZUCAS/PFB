```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my = np.loadtxt("psr.txt")
cspfb = np.loadtxt("cspfb.txt")
opfb = np.loadtxt("opfb.txt")
```

```python
my = (my - np.min(my)) / (np.max(my) - np.min(my))
cspfb = (cspfb - np.min(cspfb)) / (np.max(cspfb) - np.min(cspfb))
opfb = (opfb - np.min(opfb)) / (np.max(opfb) - np.min(opfb))
```

```python
plt.figure(figsize=(15,7))
plt.plot(my,label="my")
plt.plot(cspfb,label="cspfb")
plt.plot(opfb,label="opfb")
plt.legend()
plt.show()
```
