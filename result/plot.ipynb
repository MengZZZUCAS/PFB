```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my = np.loadtxt("psr.txt")
cspfb = np.loadtxt("cspfb.txt")
opfb = np.loadtxt("opfb.txt")
```

```python
my = (my - np.min(my)) / (np.max(my) - np.min(my))
cspfb = (cspfb - np.min(cspfb)) / (np.max(cspfb) - np.min(cspfb))
opfb = (opfb - np.min(opfb)) / (np.max(opfb) - np.min(opfb))
```

```python
index = np.linspace(0,1,1024)
```

```python
plt.figure(figsize=(15,7))
plt.plot(index,my,label="coherent_dedispersion",color = "blue")
plt.plot(index,cspfb,label="coherent_dedispersion+cspfb",color="red")
plt.plot(index,opfb,label="coherent_dedispersion+opfb",color = "k")
plt.xlabel("Phase",fontsize=15)
plt.ylabel("Relative",fontsize=15)
plt.legend()
plt.grid(alpha=1, linestyle='--')
plt.show()
```
